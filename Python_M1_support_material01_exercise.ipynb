## 4\. Booleanos

Utilize a tabela da verdade para responder: qual o valor da variável x?

```
a = False
b = True

x = not a & b
#O valor será true pois not a é true e b é true, true + true é igual a true na tabela verdade.
print("O valor da variável x é:", x)
```
